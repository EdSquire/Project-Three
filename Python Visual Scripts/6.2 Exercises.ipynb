# Categorical Plot

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(uberlyft['price'], bins = 20, kde = True)
```

```python
uberlyft.dtypes
```

```python
uberlyft.loc[uberlyft['price'] < 20.00, 'Price category'] = 'Low price'
```

```python
uberlyft.loc[(uberlyft['price'] >= 20) & (uberlyft['price'] < 40), 'Price category'] = 'Middle price'
```

```python
uberlyft.loc[uberlyft['price'] >= 40, 'Price category'] = 'High price'
```

```python
uberlyft['Price category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="distance", y="price", hue="Price category", data=uberlyft)
```

```python
# The Categorical Plot does indicate a heavy concentration and rlationship between price and distance. Additional analysis of the relationship betwen these two variables would make sense.
```
